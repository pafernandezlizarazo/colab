```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/pafernandezlizarazo/colab/main/games.csv')
df
```

```python
df.head(2)
```

```python
df1 = df[['id', 'rated', 'created_at', 'last_move_at', 'turns', 'victory_status', 'winner', 'increment_code', 'white_id', 'white_rating', 'black_id', 'black_rating', 'opening_eco', 'opening_name', 'opening_ply']]
df1
```

```python
sns.set_style('whitegrid');

text_dict = {'size':16,'color':'black'};
plt.figure(figsize=(8,6));
plt.hist(df1['black_rating'],color='red');
plt.xlabel('CalificacÃ­on de Jugador Negro',fontdict=text_dict);
```

```python
sns.set_style('whitegrid');

text_dict = {'size':16,'color':'black'};
plt.figure(figsize=(8,6));
plt.hist(df1['white_rating'],color='blue');
plt.xlabel('CalificacÃ­on de Jugador Blanco',fontdict=text_dict);
```

```python
sns.set_style('whitegrid');

text_dict = {'size':16,'color':'black'};
text_legend = {'size':16}
plt.figure(figsize=(8,6));
plt.hist(df1['black_rating'],color='red',alpha=0.9,label='J. Negro');
plt.hist(df1['white_rating'],color='blue',alpha=0.6,label='J. Blanco');
plt.xlabel('CalificacÃ­on de Jugadores',fontdict=text_dict);
plt.legend(loc='best',prop=text_legend);
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df1,x='white_rating',y='black_rating',hue='winner');
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df1,x='white_rating',y='black_rating',hue='victory_status');
```

```python
sns.jointplot(data=df1, x='white_rating',y='black_rating',color='b');
```

```python
sns.jointplot(data=df1, x='white_rating',y='black_rating',color='r',kind='hex');
plt.colorbar()
```

```python
df_mate = df1[df1['victory_status']=='mate'];
df_mate
```

```python
plt.figure(figsize=(8,6));
sns.scatterplot(data=df_mate,x='white_rating',y='black_rating',hue='winner');
```

```python
sns.jointplot(data=df_mate, x='white_rating',y='black_rating',color='b',kind='kde');
```
